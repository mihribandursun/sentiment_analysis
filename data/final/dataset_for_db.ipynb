```python
import pandas as pd

df = pd.read_csv(r"C:\Users\nurha\OneDrive\Desktop\sentiment_analysis_project\data\processed\restaurant_reviews_processed.csv")

# Restaurant table (unique by vendor_id)
restaurants_cols = [
    "vendor_id", "restaurant_name", "link", "img", "district",
    "cuisine_type", "price_range", "min_order", "delivery_time", "delivery_type", "restaurant_slug"
]
restaurants_df = df[restaurants_cols].drop_duplicates(subset=["vendor_id"])

restaurants_df.to_csv(r"C:\Users\nurha\OneDrive\Desktop\sentiment_analysis_project\data\final\restaurants.csv", index=False)

# Comment table
reviews_cols = [
    "uuid", "vendor_id", "reviewer_name", "reviewer_id",
    "overall_score", "restaurant_food_score", "rider_score",
    "text", "created_at", "like_count", "replies_count",
    "product_names", "sentiment_label", "sentiment",
    "has_rider_score", "review_length"
]
reviews_df = df[reviews_cols]

reviews_df.to_csv(r"C:\Users\nurha\OneDrive\Desktop\sentiment_analysis_project\data\final\reviews.csv", index=False)


print("Separated into restaurants and reviews.!")
print(f"Total restaurant: {len(restaurants_df)}")
print(f"Total reviews: {len(reviews_df)}")
```
